# Lasso

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
%matplotlib inline

%load_ext autoreload
% autoreload 2

n_samples, n_features = 50, 200
n_zeros = 10
X = np.random.randn(n_samples, n_features)
coef = 3 * np.random.randn(n_features)
inds = np.arange(n_features)
np.random.shuffle(inds)
coef[inds[n_zeros:]] = 0  # sparsify coef
y = np.dot(X, coef)

# add noise
y += 0.01 * np.random.normal(size=n_samples)

# Split data in train set and test set
n_samples = X.shape[0]
X_train, y_train = X[:n_samples // 2], y[:n_samples // 2]
X_test, y_test = X[n_samples // 2:], y[n_samples // 2:]
```

## sklearn Lasso

```python
from sklearn.linear_model import Lasso

alpha = 0.1
lasso = Lasso(alpha=alpha, fit_intercept=True, normalize=False)

y_pred_lasso = lasso.fit(X_train, y_train).predict(X_test)
r2_score_lasso = r2_score(y_test, y_pred_lasso)
print(lasso)
print('number of iterations: {}'.format(lasso.n_iter_))
print('Sparsity of coef: {} %'.format(np.mean(lasso.coef_ == 0) * 100))
print('Intercept: {}'.format(lasso.intercept_))
print('r^2 on test data : {}'.format(r2_score_lasso))

plt.plot(lasso.coef_, color='gold', linewidth=2,
         label='Lasso coefficients')
plt.plot(coef, '--', color='navy', label='original coefficients')
plt.legend()
```

## Coordinate Descent

```python
import coordinate_descent

alpha = 0.1
lasso = coordinate_descent.Lasso(alpha=alpha, fit_intercept=True, normalize=True)

y_pred_lasso = lasso.fit(X_train, y_train).predict(X_test)
r2_score_lasso = r2_score(y_test, y_pred_lasso)
print(lasso)
print('number of iterations: {}'.format(lasso.n_iter_))
print('Sparsity of coef: {} %'.format(np.mean(lasso.coef_ == 0) * 100))
print('Intercept: {}'.format(lasso.intercept_))
print('r^2 on test data : {}'.format(r2_score_lasso))

plt.plot(lasso.coef_, color='gold', linewidth=2,
         label='Lasso coefficients')
plt.plot(coef, '--', color='navy', label='original coefficients')
plt.legend()
```
