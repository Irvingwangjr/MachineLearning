## Generate Tree data

```python
def retrieve_tree(idx):
    list_of_trees = [{'no surfacing': {0: 'no',
                                       1: {'flippers': {0: 'no',
                                                        1: 'yes'}}}},
                     {'no surfacing': {0: 'no',
                                       1: {'flippers': {0: {'head': {0: 'no',
                                                                     1: 'yes'}},
                                                        1: 'no'}}}}]
    return list_of_trees[idx]
```
