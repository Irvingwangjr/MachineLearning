# Logistic Regression

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

## Load data

```python
raw_data = np.loadtxt('data/data1.txt', delimiter=',')
data = raw_data[:,:2]
label = raw_data[:,-1].astype(np.int)

print('Number of samples: {}, Number of features: {}'.format(data.shape[0], data.shape[1]))
print('Label range: {}'.format(np.unique(label)))
```

## Visualize data

```python
plt.plot(data[label == 0, 0], data[label == 0, 1], 'ro', label='Class 0')
plt.plot(data[label == 1, 0], data[label == 1, 1], 'yo', label='Class 1')
plt.legend()
```

## Train data

```python
from LogisticRegression import LogisticRegression

model = LogisticRegression()
model.fit(data, label)
history = np.array(model.history)

plt.plot(history, '*-')
plt.xlabel('iteration')
plt.ylabel('cost')
```

## Decision Boundary

```python
coef = model.coef_
intercept = model.intercept_

plot_x = np.array([np.amin(data[:, 0])-2, np.amax(data[:, 0])+2])
ploy_y = -1.0 / coef[1] * (coef[0] * plot_x + intercept)

plt.plot(data[label == 0, 0], data[label == 0, 1], 'ro', label='Class 0')
plt.plot(data[label == 1, 0], data[label == 1, 1], 'yo', label='Class 1')
plt.plot(plot_x, ploy_y)
plt.legend()
```
